## Numerical Differentiation Problem Statement

### Introduction

In many real-world applications, the exact mathematical function describing a system is either unknown or only available as discrete data points. When working with such data, we must rely on numerical differentiation to approximate the rate of change. This process involves discretization, where a continuous function is represented by a finite set of points on a numerical grid with a defined step size (h). Since there are only function values at specific points rather than a continuous expression, finite difference methods are used to estimate derivatives. The accuracy of these approximations depends on the choice of step size and the numerical method used. Understanding numerical differentiation and discretization is essential in fields like physics, engineering, and finance, where derivatives provide key insights into trends, rates of change, and system behavior.

<br>

### Python

In the following sections, different numerical approaches are explored to solve a problem manually, and verify the results using Python, but first, a continuous function is represented discretely using Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the numerical grid (discretization)
x = np.linspace(0, 2*np.pi, 10)  # 10 evenly spaced points from 0 to 2Ï€
y = np.sin(x)  # Function values at discrete points

# Plot the discrete points
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'bo-', label="Discrete Points")  # 'bo-' for blue circles with lines
plt.xlabel("x")
plt.ylabel("f(x) = sin(x)")
plt.title("Discrete Representation of a Continuous Function")
plt.legend()
plt.grid(True)
plt.show()
```

## Finite Difference Approximating Derivatives

### Introduction

In numerical differentiation, finite difference methods provide a way to approximate derivatives when only discrete data points are available. Since an exact function is often unknown and only sampled values are given, these methods estimate the derivative using nearby function points. The three primary finite difference methods are the forward difference, which approximates the derivative using a point and the next point; the backward difference, which relies on a point and the previous point; and the central difference, which averages the rate of change between the previous and next points, offering higher accuracy. These methods are derived from the Taylor series expansion, which expresses a function as an infinite sum of its derivatives at a given point. By truncating the Taylor series and rearranging terms, approximations for the first derivative can be obtained along with their associated error terms. The central difference method is particularly effective because its symmetric formulation cancels out lower-order error terms, making it more accurate than forward or backward differences for the same step size. The precision of these methods depends on the step size (h), where smaller values generally improve accuracy but may also introduce numerical instability due to floating-point errors.

<br>

### Python

In the next steps, finite difference approximations will be applied to compute derivatives manually and confirm the results using Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its analytical derivative
def f(x):
    return np.sin(x)

def f_prime_exact(x):
    return np.cos(x)

# Define the numerical grid
h = 0.1  # Step size
x = np.arange(0, 2*np.pi, h)  # Grid points

# Compute function values at grid points
y = f(x)

# Forward difference approximation
forward_diff = (f(x + h) - f(x)) / h

# Backward difference approximation
backward_diff = (f(x) - f(x - h)) / h

# Central difference approximation
central_diff = (f(x + h) - f(x - h)) / (2 * h)

# Compute exact derivative values
exact_derivative = f_prime_exact(x)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x, exact_derivative, label="Exact Derivative", linestyle="dashed", color="black")
plt.plot(x, forward_diff, label="Forward Difference", linestyle="dotted", color="red")
plt.plot(x, backward_diff, label="Backward Difference", linestyle="dotted", color="blue")
plt.plot(x, central_diff, label="Central Difference", linestyle="solid", color="green")
plt.xlabel("x")
plt.ylabel("Derivative Approximation")
plt.title("Finite Difference Approximations of the Derivative")
plt.legend()
plt.grid(True)
plt.show()
```

## Approximating of Higher Order Derivatives

### Introduction

Approximating higher-order derivatives numerically involves extending finite difference methods using Taylor series expansions. By expanding a function around a point $x_â€‹j$ and evaluating it at $x_{jâˆ’1}$â€‹ and $x_{j+1}$â€‹, expressions for the function values at these neighboring points can be obtained. Adding these Taylor series expansions eliminates the first derivative term, resulting in an approximation for the second derivative:

$$
\frac{fâ€²â€²(x_j)â‰ˆf(x_{j+1})âˆ’2f(x_j)+f(x_{jâˆ’1})}{h^2}
$$

This formula is derived with an $O(h^2)$ error, meaning its accuracy improves quadratically as the step size $h$ decreases. Higher-order derivatives can be approximated similarly by including additional terms from the Taylor expansion. These numerical approximations are particularly useful in solving differential equations and analyzing systems where curvature and changes in slope must be understood. The next steps will involve computing second-order derivatives numerically and verifying the results using Python.

### Python

The next steps will involve computing second-order derivatives numerically and verifying the results using Python.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its analytical second derivative
def f(x):
    return np.sin(x)

def f_double_prime_exact(x):
    return -np.sin(x)

# Define the numerical grid
h = 0.1  # Step size
x = np.arange(0, 2*np.pi, h)  # Grid points

# Compute function values at grid points
y = f(x)

# Second derivative approximation using the central difference formula
second_derivative = (f(x + h) - 2*f(x) + f(x - h)) / h**2

# Compute exact second derivative values
exact_second_derivative = f_double_prime_exact(x)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(x, exact_second_derivative, label="Exact Second Derivative", linestyle="dashed", color="black")
plt.plot(x, second_derivative, label="Numerical Approximation", linestyle="solid", color="blue")
plt.xlabel("x")
plt.ylabel("Second Derivative Approximation")
plt.title("Finite Difference Approximation of the Second Derivative")
plt.legend()
plt.grid(True)
plt.show()
```
